### **2-1. Ïã§ÏãúÍ∞Ñ Ïù¥ÎØ∏ÏßÄ Ï∫°Ï≤ò Ìï®Ïàò (`capture_photo`)**

```python
from IPython.display import display, Javascript, Image
from google.colab.output import eval_js
from base64 import b64decode, b64encode
import PIL
import io
import numpy as np
```

#### **Ïù¥ÎØ∏ÏßÄ Ï∫°Ï≤ò Ìï®Ïàò Ï†ïÏùò**

Ïù¥Ï†ú Î≥∏Í≤©Ï†ÅÏúºÎ°ú ÏõπÏ∫†ÏùÑ Ï†úÏñ¥ÌïòÍ≥† ÏÇ¨ÏßÑÏùÑ Ï∞çÎäî `capture_photo` Ìï®ÏàòÎ•º ÎßåÎì≠ÎãàÎã§. Ìï®ÏàòÏùò ÎÇ¥Î∂Ä ÎèôÏûëÏùÑ Îã®Í≥ÑÎ≥ÑÎ°ú ÏÇ¥Ìé¥Î≥¥ÏÑ∏Ïöî.

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode
import cv2
import numpy as np
from datetime import datetime

def capture_photo(quality=0.8):
  """
  Google Colab ÌôòÍ≤ΩÏóêÏÑú ÏõπÏ∫†ÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Ïù¥ÎØ∏ÏßÄÎ•º Ï∫°Ï≤òÌïòÍ≥† Ï†ÄÏû•Ìï©ÎãàÎã§. (ÏàòÏ†ïÎêú Î≤ÑÏ†Ñ)
  """
  # ÏûêÎ∞îÏä§ÌÅ¨Î¶ΩÌä∏ ÏΩîÎìúÎ•º ÌïòÎÇòÏùò Î¨∏ÏûêÏó¥Î°ú ÌÜµÌï© (Ìï®Ïàò Ï†ïÏùò + Ïã§Ìñâ)
  js_code = f"""
    async function takePhoto(quality) {{
      // 1. UI ÏöîÏÜå ÏÉùÏÑ± (ÎπÑÎîîÏò§, Î≤ÑÌäº Îì±)
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Ï∫°Ï≤ò';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({{video: true}});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // ColabÏùò Ï∂úÎ†• ÌîÑÎ†àÏûÑ ÌÅ¨Í∏∞Ïóê ÎßûÍ≤å UI Ï°∞Ï†ï
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // 2. Ï∫°Ï≤ò Î≤ÑÌäº ÌÅ¥Î¶≠ ÎåÄÍ∏∞
      await new Promise((resolve) => capture.onclick = resolve);

      // 3. Ï∫îÎ≤ÑÏä§Ïóê ÌòÑÏû¨ ÌîÑÎ†àÏûÑ Í∑∏Î¶¨Í∏∞
      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);

      // 4. Ïä§Ìä∏Î¶º Î∞è UI Ï†ïÎ¶¨
      stream.getTracks().forEach(track => track.stop());
      document.body.removeChild(div);

      // 5. Ïù¥ÎØ∏ÏßÄ Îç∞Ïù¥ÌÑ∞Î•º Base64Î°ú Ïù∏ÏΩîÎî©ÌïòÏó¨ Î∞òÌôò
      return canvas.toDataURL('image/jpeg', quality);
    }}
    // Ï†ïÏùòÌïú Ìï®ÏàòÎ•º Î∞îÎ°ú Ïã§ÌñâÌïòÏó¨ Í≤∞Í≥ºÎ•º Î∞òÌôò
    takePhoto({quality});
  """

  try:
    # ÏúÑÏóêÏÑú Ï†ïÏùòÌïòÍ≥† Ïã§ÌñâÌïú ÏûêÎ∞îÏä§ÌÅ¨Î¶ΩÌä∏ ÏΩîÎìúÏùò Í≤∞Í≥ºÎ•º Î∞õÏùå
    data_url = eval_js(js_code)

    # Îç∞Ïù¥ÌÑ∞ URL ÎîîÏΩîÎî© (Ìó§Îçî Ï†úÍ±∞, Base64 ÎîîÏΩîÎî©)
    binary_data = b64decode(data_url.split(',')[1])

    # OpenCVÏóêÏÑú ÏÇ¨Ïö©Ìï† Ïàò ÏûàÎäî numpy Î∞∞Ïó¥Î°ú Î≥ÄÌôò
    nparr = np.frombuffer(binary_data, np.uint8)
    img = cv2.imdecode(nparr, cv2.IMREAD_COLOR)

    # ÌååÏùºÎ°ú Ï†ÄÏû•
    filename = f"capture_{datetime.now().strftime('%Y%m%d_%H%M%S')}.jpg"
    cv2.imwrite(filename, img)

    print(f"‚úÖ Ïù¥ÎØ∏ÏßÄÍ∞Ä '{filename}' ÌååÏùºÎ°ú Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§.")
    return filename
  except Exception as err:
    # eval_jsÏóêÏÑú Ïò§Î•òÍ∞Ä Î∞úÏÉùÌïòÎ©¥ Ïù¥ Î∂ÄÎ∂ÑÏù¥ Ïã§ÌñâÎê©ÎãàÎã§.
    print(f"Ïò§Î•ò: ÏõπÏ∫†ÏùÑ ÏÇ¨Ïö©Ìï† Ïàò ÏóÜÏäµÎãàÎã§. ({err})")
    print("Î∏åÎùºÏö∞Ï†ÄÏóêÏÑú Ïπ¥Î©îÎùº Ï†ëÍ∑º Í∂åÌïúÏùÑ ÌóàÏö©ÌñàÎäîÏßÄ ÌôïÏù∏Ìï¥Ï£ºÏÑ∏Ïöî.")
    return None

# --- Ìï®Ïàò Ïã§Ìñâ Î∞è Í≤∞Í≥º ÌôïÏù∏ ---
image_file = capture_photo()

# Ï†ÄÏû•Îêú Ïù¥ÎØ∏ÏßÄ ÌôïÏù∏ (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
if image_file:
  from IPython.display import Image
  display(Image(image_file))
```

### **2-2. Ïã§ÏãúÍ∞Ñ ÏùåÏÑ± ÎÖπÏùå Ìï®Ïàò (`record_audio`)**

Ïù¥Î≤àÏóêÎäî ÎßàÏù¥ÌÅ¨Î°ú ÏùåÏÑ±ÏùÑ ÎÖπÏùåÌïòÎäî Í∏∞Îä•ÏùÑ ÎßåÎì§Í≤†ÏäµÎãàÎã§. Ïù¥ÎØ∏ÏßÄ Ï∫°Ï≤òÏôÄ Í±∞Ïùò ÎèôÏùºÌïú ÏõêÎ¶¨Î°ú ÏûëÎèôÌï©ÎãàÎã§.

#### **ÌïÑÏöîÌïú ÎèÑÍµ¨(ÎùºÏù¥Î∏åÎü¨Î¶¨) Í∞ÄÏ†∏Ïò§Í∏∞**

Ïò§ÎîîÏò§ Ï≤òÎ¶¨Î•º ÏúÑÌï¥ `pydub`Ïù¥ÎùºÎäî ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º Ï∂îÍ∞ÄÎ°ú ÏÑ§ÏπòÌïòÍ≥† ÏÇ¨Ïö©Ìï©ÎãàÎã§. Î∏åÎùºÏö∞Ï†ÄÍ∞Ä ÎÖπÏùåÌïòÎäî Ïò§ÎîîÏò§ ÌòïÏãù(WebM)ÏùÑ AIÍ∞Ä Îçî Ïûò Ïù¥Ìï¥ÌïòÎäî ÌëúÏ§Ä ÌòïÏãù(WAV)ÏúºÎ°ú Î≥ÄÌôòÌïòÎäî Îç∞ Íº≠ ÌïÑÏöîÌï©ÎãàÎã§.


```python
!pip install pydub -q

from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode
import pydub
import io
from datetime import datetime

def record_audio(filename_prefix="audio_record"):
  """
  Google Colab ÌôòÍ≤ΩÏóêÏÑú ÎßàÏù¥ÌÅ¨Î•º ÏÇ¨Ïö©ÌïòÏó¨ Ïò§ÎîîÏò§Î•º ÎÖπÏùåÌïòÍ≥† WAV ÌååÏùºÎ°ú Ï†ÄÏû•Ìï©ÎãàÎã§. (ÏàòÏ†ïÎêú Î≤ÑÏ†Ñ)
  """
  # ÏûêÎ∞îÏä§ÌÅ¨Î¶ΩÌä∏ ÏΩîÎìúÎ•º ÌïòÎÇòÏùò Î¨∏ÏûêÏó¥Î°ú ÌÜµÌï©
  js_code = """
    async function recordAudio() {
      // 1. UI ÏöîÏÜå ÏÉùÏÑ± (Î≤ÑÌäº)
      const div = document.createElement('div');
      const startButton = document.createElement('button');
      startButton.textContent = 'ÎÖπÏùå ÏãúÏûë';
      div.appendChild(startButton);

      const stopButton = document.createElement('button');
      stopButton.textContent = 'ÎÖπÏùå Ï§ëÏßÄ';
      stopButton.disabled = true;
      div.appendChild(stopButton);

      const status = document.createElement('p');
      status.textContent = 'Î≤ÑÌäºÏùÑ ÎàåÎü¨ ÎÖπÏùåÏùÑ ÏãúÏûëÌïòÏÑ∏Ïöî.';
      div.appendChild(status);

      // Colab Ï∂úÎ†•Ï∞ΩÏóê UI ÌëúÏãú
      document.body.appendChild(div);
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // 2. ÎÖπÏùå Í¥ÄÎ†® ÏÑ§Ï†ï
      const stream = await navigator.mediaDevices.getUserMedia({ audio: true });
      const mediaRecorder = new MediaRecorder(stream);
      let audioChunks = [];

      mediaRecorder.ondataavailable = event => {
        audioChunks.push(event.data);
      };

      // 3. ÎÖπÏùå ÏãúÏûë/Ï§ëÏßÄ Î°úÏßÅ
      const startPromise = new Promise(resolve => {
        startButton.onclick = () => {
          mediaRecorder.start();
          startButton.disabled = true;
          stopButton.disabled = false;
          status.textContent = 'ÎÖπÏùå Ï§ë...';
          resolve();
        };
      });

      const stopPromise = new Promise(resolve => {
        stopButton.onclick = () => {
          mediaRecorder.stop();
          stopButton.disabled = true;
          status.textContent = 'ÎÖπÏùå ÏôÑÎ£å. Ï≤òÎ¶¨ Ï§ë...';
          resolve();
        };
      });

      await startPromise;
      await stopPromise;

      // 4. ÎÖπÏùå Îç∞Ïù¥ÌÑ∞ Ï≤òÎ¶¨ Î∞è Î∞òÌôò
      return new Promise(resolve => {
        mediaRecorder.onstop = () => {
          const audioBlob = new Blob(audioChunks, { type: 'audio/webm' });
          const reader = new FileReader();
          reader.readAsDataURL(audioBlob);
          reader.onloadend = () => {
            const base64data = reader.result;
            // Ïä§Ìä∏Î¶º Î∞è UI Ï†ïÎ¶¨
            stream.getTracks().forEach(track => track.stop());
            document.body.removeChild(div);
            resolve(base64data);
          };
        };
      });
    }
    // Ï†ïÏùòÌïú Ìï®ÏàòÎ•º Î∞îÎ°ú Ïã§ÌñâÌïòÏó¨ Í≤∞Í≥ºÎ•º Î∞òÌôò
    recordAudio();
  """

  try:
    # ÏúÑÏóêÏÑú Ï†ïÏùòÌïòÍ≥† Ïã§ÌñâÌïú ÏûêÎ∞îÏä§ÌÅ¨Î¶ΩÌä∏ ÏΩîÎìúÏùò Í≤∞Í≥ºÎ•º Î∞õÏùå
    data_url = eval_js(js_code)

    # Îç∞Ïù¥ÌÑ∞ URL ÎîîÏΩîÎî©
    binary_data = b64decode(data_url.split(',')[1])

    # Î∏åÎùºÏö∞Ï†ÄÎäî Î≥¥ÌÜµ webm/opus Ìè¨Îß∑ÏúºÎ°ú ÎÖπÏùåÌïòÎØÄÎ°ú, pydubÏùÑ ÏÇ¨Ïö©Ìï¥ wavÎ°ú Î≥ÄÌôò
    audio_segment = pydub.AudioSegment.from_file(io.BytesIO(binary_data), format="webm")

    # ÌååÏùºÎ°ú Ï†ÄÏû•
    filename = f"{filename_prefix}_{datetime.now().strftime('%Y%m%d_%H%M%S')}.wav"
    audio_segment.export(filename, format="wav")

    print(f"‚úÖ Ïò§ÎîîÏò§Í∞Ä '{filename}' ÌååÏùºÎ°ú Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§.")
    return filename
  except Exception as err:
    # eval_jsÏóêÏÑú Ïò§Î•òÍ∞Ä Î∞úÏÉùÌïòÎ©¥ Ïù¥ Î∂ÄÎ∂ÑÏù¥ Ïã§ÌñâÎê©ÎãàÎã§.
    print(f"Ïò§Î•ò: ÎßàÏù¥ÌÅ¨Î•º ÏÇ¨Ïö©Ìï† Ïàò ÏóÜÏäµÎãàÎã§. ({err})")
    print("Î∏åÎùºÏö∞Ï†ÄÏóêÏÑú ÎßàÏù¥ÌÅ¨ Ï†ëÍ∑º Í∂åÌïúÏùÑ ÌóàÏö©ÌñàÎäîÏßÄ ÌôïÏù∏Ìï¥Ï£ºÏÑ∏Ïöî.")
    return None

# --- Ìï®Ïàò Ïã§Ìñâ Î∞è Í≤∞Í≥º ÌôïÏù∏ ---
audio_file = record_audio()

# Ï†ÄÏû•Îêú Ïò§ÎîîÏò§ Ïû¨ÏÉù (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
if audio_file:
  from IPython.display import Audio
  display(Audio(audio_file, autoplay=False))
```

    *   `record_microphone_to_pcm16k` Ìï®ÏàòÎäî Î∏åÎùºÏö∞Ï†ÄÎ°úÎ∂ÄÌÑ∞ Ï†ÑÎã¨Î∞õÏùÄ WebM Ïò§ÎîîÏò§Î•º AI Î™®Îç∏Ïù¥ Í∞ÄÏû• ÏÑ†Ìò∏ÌïòÎäî ÌòïÏãùÏúºÎ°ú Î≥ÄÌôòÌïòÎäî 'Ï†ÑÎ¨∏ Î≤àÏó≠Í∞Ä'ÏûÖÎãàÎã§.

    *   Ïù¥Î†áÍ≤å Î≥ÄÌôòÎêú ÏàúÏàòÌïú Ïò§ÎîîÏò§ Îç∞Ïù¥ÌÑ∞Îäî ÌååÏùºÎ°ú Ï†ÄÏû•ÎêòÏßÄ ÏïäÍ≥†, Îã§Ïùå ÏÖÄÏóêÏÑú Î∞îÎ°ú Gemini ÏÑúÎ≤ÑÎ°ú Ï†ÑÏÜ°Îê† Ï§ÄÎπÑÎ•º ÎßàÏπ©ÎãàÎã§.

Ïù¥ ÏÖÄÏùÑ Ïã§ÌñâÌïòÎ©¥ "üéõÔ∏è Ï§ÄÎπÑ ÏôÑÎ£å"ÎùºÎäî Î©îÏãúÏßÄÍ∞Ä ÎÇòÌÉÄÎÇòÎ©∞, Îã§Ïùå Îã®Í≥ÑÏóêÏÑú Ïã§ÏãúÍ∞Ñ ÎåÄÌôîÎ•º ÏãúÏûëÌï† Î™®Îì† Í∏∞Ïà†Ï†ÅÏù∏ Ï§ÄÎπÑÍ∞Ä ÏôÑÎ£åÎêòÏóàÏùåÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§.

```python
# --- ÏÖÄ 1: ÎßàÏù¥ÌÅ¨ ÎÖπÏùå UI(JS) + webm‚ÜíPCM16k Î≥ÄÌôò ---
from IPython.display import Javascript, display
from google.colab.output import eval_js
from pydub import AudioSegment

js_record = r"""
async function recordAudioUI() {
  const div = document.createElement('div');
  div.style.margin = '8px 0';
  const startBtn = document.createElement('button');
  const stopBtn  = document.createElement('button');
  const status   = document.createElement('p');
  startBtn.textContent = 'üéôÔ∏è ÎÖπÏùå ÏãúÏûë';
  stopBtn.textContent  = '‚èπÔ∏è ÎÖπÏùå Ï§ëÏßÄ';
  stopBtn.disabled = true;
  status.textContent = 'Î≤ÑÌäºÏùÑ ÎàåÎü¨ ÎÖπÏùåÏùÑ ÏãúÏûëÌïòÏÑ∏Ïöî.';
  status.style.margin = '8px 0 0 0';
  div.appendChild(startBtn); div.appendChild(stopBtn); div.appendChild(status);
  document.body.appendChild(div);
  google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

  const stream = await navigator.mediaDevices.getUserMedia({audio:true});
  const rec = new MediaRecorder(stream, {mimeType: 'audio/webm'});
  const chunks = [];
  rec.ondataavailable = e => { if (e.data.size > 0) chunks.push(e.data); };

  await new Promise(resolve => {
    startBtn.onclick = () => {
      rec.start();
      startBtn.disabled = true; stopBtn.disabled = false;
      status.textContent = 'ÎÖπÏùå Ï§ë... ÎßêÌïòÏÑ∏Ïöî.';
    };
    stopBtn.onclick = () => {
      rec.stop(); stopBtn.disabled = true;
      status.textContent = 'Ï≤òÎ¶¨ Ï§ë...';
      resolve();
    };
  });

  const base64 = await new Promise(resolve => {
    rec.onstop = async () => {
      stream.getTracks().forEach(t => t.stop());
      const blob = new Blob(chunks, {type: 'audio/webm'});
      const reader = new FileReader();
      reader.onloadend = () => {
        document.body.removeChild(div);
        resolve(reader.result); // data:...;base64,XXXX
      };
      reader.readAsDataURL(blob);
    };
  });

  return base64;
}
recordAudioUI();
"""

def record_microphone_to_pcm16k():
  """Î∏åÎùºÏö∞Ï†ÄÏóêÏÑú ÎÖπÏùå(webm/opus) ‚Üí 16kHz Î™®ÎÖ∏ 16-bit PCM Î∞îÏù¥Ìä∏Î°ú Î≥ÄÌôò"""
  data_url = eval_js(js_record)               # ‚Üê Î¨∏ÏûêÏó¥ Í∑∏ÎåÄÎ°ú ÎÑòÍ≤®Ïïº Ìï® (Javascript(...) Í∞ùÏ≤¥ X)
  if not isinstance(data_url, str) or not data_url.startswith("data:"):
      raise RuntimeError("Î∏åÎùºÏö∞Ï†Ä ÎÖπÏùå Í≤∞Í≥ºÍ∞Ä ÎπÑÏ†ïÏÉÅÏûÖÎãàÎã§. ÎßàÏù¥ÌÅ¨ Í∂åÌïú/Î∏åÎùºÏö∞Ï†Ä ÏΩòÏÜîÏùÑ ÌôïÏù∏ÌïòÏÑ∏Ïöî.")
  b64 = data_url.split(",")[1]
  webm_bytes = base64.b64decode(b64)          # ‚Üê b64decode (b6decode Ïò§ÌÉÄ Í∏àÏßÄ!)
  seg = AudioSegment.from_file(io.BytesIO(webm_bytes), format="webm")  # ffmpeg ÌïÑÏöî
  seg = seg.set_frame_rate(16000).set_channels(1).set_sample_width(2)  # 16kHz, mono, 16-bit
  return seg.raw_data  # WAV Ìó§Îçî ÏóÜÏù¥ ÏàúÏàò PCM Î∞îÏù¥Ìä∏

print("üéõÔ∏è Ï§ÄÎπÑ ÏôÑÎ£å: Îã§Ïùå ÏÖÄÏóêÏÑú ÎÖπÏùå‚ÜíÎ™®Îç∏ ÏùëÎãµÍπåÏßÄ Ïã§ÌñâÌï©ÎãàÎã§.")
```
